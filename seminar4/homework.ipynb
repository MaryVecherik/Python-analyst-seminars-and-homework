# Задача 2.
### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график\
Сделайте выводы

```python
plt.figure(figsize=(8, 6))

memory = df['waterfront'].value_counts()

plt.pie(memory, autopct='%1.1f%%')
plt.title('Доля домов с видом на набережную')
plt.legend(['нет вида', 'есть вид']);
```

Расппределение показало, что только 0.8 % домов имеют вид на набережную.

### 2.2 Изучите распределение этажей домов
Постройте график\
Сделайте выводы

```python
plt.figure(figsize=(8, 6))

memory = df['floors'].value_counts()
sns.barplot(x = memory.index, y = memory)

plt.title('Распределение этажей домов')
plt.xlabel('Этаж')
plt.ylabel('Кол-во');
```

Распределение показало, что в основном дома имеют 1 и 2 этажа.

### 2.2 Изучите распределение состояния домов
Постройте график\
Сделайте выводы

```python
plt.figure(figsize=(8, 6))

memory = df['condition'].value_counts()
sns.barplot(x = memory.index, y = memory)

plt.title('Распределение состояния домов')
plt.xlabel('Оценка состояния дома')
plt.ylabel('Кол-во');
```

Распределение показало, что большинство домов оценены в 3 балла

# Задача 3.
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.\
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.drop(columns=['id', 'date', 'zipcode'], inplace=True)
df.head()
```

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Корреляция');
```

```python
# Влияние жилой площади на стоимость
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Цена дома напрямую зависит от квадратных метров жилой площади

```python
# Влияет ли год постройки дома на его стоимость?
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg');
```

```python
# Влияние количества этажей на стоимость
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.title('Влияние количества этажей на стоимость');
```

Наибольшую стоимость имеют дома, где кол-во этажей равно 2,5 

```python
# Влияние оценки недвижимости на стоимость
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.title('Влияние оценки недвижимости на стоимость');
```
