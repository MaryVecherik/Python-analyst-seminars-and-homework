```python
df = pd.read_csv('laptop_price.csv', sep=',', encoding='latin-1')
df.head()
```

### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
# Вариант 1
# def f(x :str): 
#     return x.split()[0]

# df['Cpu_Company'] = df['Cpu'].apply(f)
# df.head()

# Вариант 2
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df.head()
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
def f(x :str):
    x = x.split()
    res = 0
    for i in x:
        if i.endswith('GB') and i[:-2].isdigit():
            res += int(i[:-2])
        elif i.endswith('TB') and i[:-2].isdigit():
            res += int(i[:-2]) * 1024
    return res

df['Memory_Amount'] = df['Memory'].apply(f)
df.head()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др)

```python
def f(x :str):
    res = []
    if 'HDD' in x:
        res.append('HDD')
    if 'SSD' in x:
        res.append('SSD')
    if not res:
        return 'other'
    return '/'.join(res)

df['Memory_Type'] = df['Memory'].apply(f)
df.head()

# Проверка
# list = ['128GB SSD', '128GB Flash Storage', '256GB SSD', '128GB SSD + 1TB HDD']
# print(f'{f(list[3])}')
```

### 1.4 Удалить признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```

```python
df['SSD'] = 0
df.head()
```

### 2.2 Поставьте в признаке SDD 1, если ноутбук действительно с типом носителя SDD

```python
df['SSD'] = df['Memory_Type'].apply(lambda x: 1 if 'SSD' in x else 0)
df.head()
```

```python
df['Weight'] = df['Weight'].str.replace('kg', '')
df['Weight'] = df['Weight'].astype(float)
```

# Задача 4.
Составьте несколько сводных таблиц
### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
Отсортируйте от меньшей стоимости к большей

```python
grouped = df.groupby('Company')
grouped.agg({'Price_euros': 'mean'}).sort_values('Price_euros')
```

### 4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора

```python
grouped = df.groupby('Cpu_Company')
grouped.agg({'Price_euros': ['min', 'mean', 'max']}).sort_values(('Price_euros', 'mean'))
```

### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ

```python
df.pivot_table(index='Cpu_Company', columns='Ram', aggfunc='count', values='Product')   
```

### 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти

```python
df.pivot_table(index='OpSys', columns='Memory_Amount', aggfunc='mean', values='Price_euros', fill_value=0)  
```
